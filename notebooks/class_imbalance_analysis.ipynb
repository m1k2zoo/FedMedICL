# Analysis of Class Imbalance in Medical Imaging Datasets

This notebook is designed to provide a comprehensive analysis of class imbalance across the medical imaging datasets in FedMedICL. Class imbalance is a common issue in machine learning for healthcare, where some classes (or labels) are underrepresented compared to others. This can significantly affect the performance of machine learning models, making them less effective at detecting rare conditions.

Throughout this notebook, we will:
- Load and inspect dataset configurations.
- Visualize the distribution of classes within multiple datasets such as 'OL3I', 'fitzpatrick17k', 'PAPILA', 'HAM10000', and 'CheXpert'.
- Analyze how class distribution varies with respect to different binary features like age or skin type.

By the end of this notebook, you will have a clear understanding of the significant class imbalance in various medical imaging datasets.

### Library imports and path management

```python
import json
import os
from os.path import dirname as up
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set the root directory and load the configuration file that contains paths to the datasets.
root_dir = up(os.getcwd())
with open(root_dir + "/configs/datasets.json", 'r') as f:
    config = json.load(f)
```

### We first load the metadata for the 'OL3I' dataset. 

This metadata contains details necessary for analyzing the dataset, such as labels or classes associated with each image.

```python
dataset_name = "OL3I"
dataset_config = config[dataset_name]
train_meta = pd.read_csv(dataset_config['train_meta_path']) 
val_meta = pd.read_csv(dataset_config['val_meta_path']) 
test_meta = pd.read_csv(dataset_config['test_meta_path']) 
```

### Visual analysis of the class distribution in the 'OL3I' dataset. 

We create a set of histograms that display the frequency of each label in both the training and test sets. The visualization clearly shows the severe class imbalances, which can bias model training.

```python

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

# Plot histograms on each subplot
train_meta['Target'].value_counts().plot(kind='bar', ax=axs[0], title='Labels of train')
test_meta['Target'].value_counts().plot(kind='bar', ax=axs[1], title='Labels of test')

# Adjust subplot layout
plt.tight_layout()

# Show the plot
plt.show()
```

### A deeper look into class imbalance in old vs young patients.
To delve deeper into the class imbalance, we analyze how labels distribute across different groups defined by binary features such as age. This analysis helps us understand if the imbalance is consistent across different demographic or clinical characteristics.

```python
# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# Plot histograms on each subplot
train_meta.loc[train_meta["Age_binary"] == 0]['Target'].value_counts().plot(kind='bar', ax=axs[0, 0], title='Labels of Group 0 (train set)')
train_meta.loc[train_meta["Age_binary"] == 1]['Target'].value_counts().plot(kind='bar', ax=axs[0, 1], title='Labels of Group 1 (train set)')
test_meta.loc[train_meta["Age_binary"] == 0]['Target'].value_counts().plot(kind='bar', ax=axs[1, 0], title='Labels of Group 0 (test set)')
test_meta.loc[train_meta["Age_binary"] == 1]['Target'].value_counts().plot(kind='bar', ax=axs[1, 1], title='Labels of Group 1 (test set)')

# Adjust subplot layout
plt.tight_layout()

# Show the plot
plt.show()
```
